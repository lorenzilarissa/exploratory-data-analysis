In this code, I've started by importing the necessary libraries: numpy, pandas, matplotlib.pyplot, seaborn, and scipy.stats.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

I then create a random dataset using numpy.random.normal().

```python
# Create a random dataset
np.random.seed(42)
data = np.random.normal(loc=5, scale=2, size=100)
```

Next, I've calculated basic statistics such as mean, median, mode, standard deviation, and variance.

```python
# Basic statistics
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=False)[0]
std_dev = np.std(data)
variance = np.var(data)
```

I've ploted a histogram of the data using seaborn.histplot() and added vertical lines to represent the mean, median, and mode using matplotlib.pyplot.axvline().

```python
# Histogram plot
plt.figure(figsize=(8, 6))
sns.histplot(data, kde=True)
plt.axvline(mean, color='r', linestyle='--', label='Mean')
plt.axvline(median, color='g', linestyle='--', label='Median')
plt.axvline(mode, color='b', linestyle='--', label='Mode')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with Mean, Median, and Mode')
plt.legend()
plt.show()
```

I've also created a box plot using seaborn.boxplot() to visualize the distribution of the data and a quantile-quantile (Q-Q) plot using scipy.stats.probplot() to assess the normality of the data.

```python
# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

```python
# Quantile-quantile (Q-Q) plot
plt.figure(figsize=(8, 6))
stats.probplot(data, dist='norm', plot=plt)
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.title('Q-Q Plot')
plt.show()
```

I've performed a t-test using scipy.stats.ttest_1samp() to test if the mean of the data is significantly different from a specified value (in this case, 5). And then I've printed the t-statistic and p-value, and based on the p-value, I've determined whether to reject or not reject the null hypothesis.

```python
# Statistical analysis
t_stat, p_value = stats.ttest_1samp(data, 5)
print(f'T-Statistic: {t_stat:.3f}')
print(f'P-Value: {p_value:.3f}')
if p_value < 0.05:
    print('The null hypothesis is rejected.')
else:
    print('The null hypothesis is not rejected.')
```

For correlation analysis, I've created two random variables x and y and calculated the correlation coefficient using numpy.corrcoef().

```python
# Correlation analysis
np.random.seed(42)
x = np.random.normal(loc=5, scale=2, size=100)
y = x + np.random.normal(loc=0, scale=1, size=100)
correlation = np.corrcoef(x, y)[0, 1]
print(f'Correlation coefficient: {correlation:.3f}')
```

I then ploted a scatter plot using matplotlib.pyplot.scatter() to visualize the relationship between x and y.

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```
